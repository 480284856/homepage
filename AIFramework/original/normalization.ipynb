Why do we need Batch Normalization?

Imagine a double-layer neural network where the input to the first layer is fixed. For example, if we feed the entire dataset into the network at once, the data distribution remains the same for each forward computation(or for the first linear layer). If the parameters of the first linear layer keep the same, the output distribution of this linear layer also remains unchanged and fixed.



However, once we update the parameters, the output distribution changes. It shifts from one place to another with every parameter update, for the bottom linear layer. This means its output distribution is constantly shifting.



This creates a problem for the second linear layer. Because its input distribution is inconsistent, for example, it might see data centered around a mean of 0 and variance of 1 the first time, move to around -1 the second time, and shift to around -5 the third time.

Consequently, it becomes very difficult for subsequent layers (the second layer here) to learn anything useful. Unlike the first layer, which receives the entire dataset with a constant input distribution, **the second layer's input is always changing**!

This is where Batch Normalization comes in. It functions by taking the shifting output distributions from a linear layer and changing them into a fixed distribution. This ensures that the linear layers(the second layer here) using the previous network's output as their input receive a stable signal, which ultimately allows for better learning.
